### Creating a Config File for Use with OCI CLI


A configuration file is needed to utilize OCI SDKs and CLI. This file contains information like user credentials and tenancy OCID. This notebook is an example of how to create a configuration file to use when accessing files from Oracle Open Data using CLI to access object storage. 

First, navigte to the base directory you would like your oci files to be located in. The default configuration location for the config file is `~/.oci/config`. To setup your config file, import oci and os, and then run the command `oci setup config` in the terminal. This will walk you through a series of steps to create your config file. The user and tenancy OCIDs you will need during these steps can be found in the OCI console. 

```python
import oci
import os

!oci setup config
```

Follow [these](https://docs.oracle.com/en-us/iaas/Content/API/Concepts/apisigningkey.htm#How2) instructions. You can use the `cat` command to view and copy the contents of the public key console, which you should then upload to the associated user in the console. 

```python
!cd .oci
!cat oci_api_key_public.pem
```

You can add the public key contents to you user by accessing the user settings menu in the [OCI console](https://www.oracle.com/cloud/sign-in.html?redirect_uri=https%3A%2F%2Fcloud.oracle.com%2F) and navigating to API keys and selecting the option to either upload your key or paste the contents. 

After adding your key, you should see a configuration file preview box. Paste the contents of the text box into your ~/.oci/config file and update the key_file parameter with the file path to your private key.
